```python
from datetime import datetime

import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

import plotly.plotly as py
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
import plotly.graph_objs as go
from plotly import tools
```

```python
def convert():
    orig = open('./env_wasgen_1_Data.csv', 'r')
    text = orig.read()
    text = text.replace('Year', 'year')
    text = text.replace('WASTE', 'waste')
    text = text.replace('GEO', 'country')
    text = text.replace('Value', 'value')
    text = text.replace('European Union (current composition)', 'EU')

    output = open('./env_wasgen_1_Data.csv', 'w')
    output.write(text)
    orig.close()
    output.close()
    
convert()
```

```python
datafile = './env_wasgen_1_Data.csv'
df = pd.read_csv(datafile)
df = df.set_index(['waste'])
df = df.loc['Plastic wastes']

# df = df.set_index(['HAZARD'])
# df = df.loc['Non-hazardous']

# df = df.set_index(['value'])
# df =df.applymap(lambda x: str(x.replace(',','.')))

df.head()
```

```python
df_2008 = df['year']==2010
df_2008 = df['country'] != 'EU'
df_2008 = df[df_2008]
df_2008.head()
df_2008 = df_2008.set_index(['year'])

df_2008
```

```python
colors=[[0, 'rgb(102,194,165)'], [0.05, 'rgb(102,194,165)'], 
              [0.15, 'rgb(171,221,164)'], [0.2, 'rgb(230,245,152)'], 
              [0.25, 'rgb(255,255,191)'], [0.35, 'rgb(254,224,139)'], 
              [0.45, 'rgb(253,174,97)'], [0.55, 'rgb(213,62,79)'], [1.0, 'rgb(158,1,66)']]

plotmap = [ dict(
        type = 'choropleth',
        locations = df_2008['country'],
        locationmode = 'country names',
        z = df_2008['value'],
        text = df_2008['value'],
        colorscale = colors,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            title = ""),
      ) ]

layout = dict(
    title = "Plastic wates in the UE countries",
    geo = dict(
        showframe = False,
        showcoastlines = True,
        showocean = True,
        oceancolor = '#26466D',
        projection = dict(
            type = 'orthographic'
        )
    ),
    height=700,
    width=900
)

fig = dict( data=plotmap, layout= layout )
iplot(fig)
```
